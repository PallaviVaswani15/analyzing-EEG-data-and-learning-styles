# Comparing Interevention and preferences of subjects for Emotiv


```python
import pandas as pd
# Load the CSV file into a DataFrame
df = pd.read_csv('merged_output.csv')
df
```

```python
# Create a function to check if any Intervention is in Preferences
def check_intervention_in_preferences(row):
    preferences = set(row['Preference'].split('|'))
    intervention = set(row['Intervention'].split('|'))
    return any(item in preferences for item in intervention)

# Filter rows where EEG-device is either "Emotiv" or "openBCI"
eeg_filtered_df = df[df['EEG-device'].isin(['Emotiv', 'openBCI'])]

# Create a new column 'match' to check if Intervention is in Preferences
eeg_filtered_df['match'] = eeg_filtered_df.apply(check_intervention_in_preferences, axis=1)

matching_subjects = eeg_filtered_df[eeg_filtered_df['match']]

# Filter the DataFrame to get subjects with matching interventions and preferences
matching_subjects['EEG-device'] = eeg_filtered_df['EEG-device']
matching_subjects['Preference'] = eeg_filtered_df['Preference']
matching_subjects['Intervention'] = eeg_filtered_df['Intervention']

# Print the names of the matching subjects along with the EEG device, Preferences, and Intervention
print(matching_subjects[['Subjects', 'EEG-device', 'Preference', 'Intervention']])
```

```python
# Filter the DataFrame to get subjects with EEG-device "Emotiv" or "openBCI" and where intervention and preferences do not match
non_matching_subjects = eeg_filtered_df[~eeg_filtered_df['match']]

# Display the information of non-matching subjects
print(non_matching_subjects[['Subjects', 'EEG-device', 'Preference', 'Intervention']])
```

```python
# Combine matching and non-matching subjects DataFrames
combined_subjects = pd.concat([matching_subjects, non_matching_subjects])

# Filter subjects with Emotiv device and similar interventions
emotiv_similar_intervention = combined_subjects[
    (combined_subjects['EEG-device'] == 'Emotiv') &
    combined_subjects.duplicated(subset='Intervention', keep=False)
]

# Print subjects with Emotiv device, similar interventions, and their subject, intervention, and preferences
print("Subjects with Emotiv Device and Similar Interventions:")
print(emotiv_similar_intervention[['Subjects', 'EEG-device', 'Intervention', 'Preference']])
```

```python
#Normality check
from scipy.stats import shapiro

# Function to check normality for a single column
def check_normality(data, column_name):
    # Check if the column contains numeric data
    if pd.api.types.is_numeric_dtype(data[column_name]):
        stat, p_value = shapiro(data[column_name])
        return stat, p_value
    else:
        return None, None

# Check normality for all numeric columns in s32_df
print("Normality test for s32_df:")
for column in s32_df.columns:
    stat, p_value = check_normality(s32_df, column)
    if stat is not None and p_value is not None:
        print(f'Column: {column}, Statistic: {stat}, p-value: {p_value}')

# Check normality for all numeric columns in s33_df
print("\nNormality test for s33_df:")
for column in s33_df.columns:
    stat, p_value = check_normality(s33_df, column)
    if stat is not None and p_value is not None:
        print(f'Column: {column}, Statistic: {stat}, p-value: {p_value}')

```

The p-values for the normality tests are extremely close to zero, indicating that the data in both s32_df and s33_df columns do not follow a normal distribution. Since the p-values are less than any reasonable significance level (e.g., 0.05), we can conclude that the data is not normally distributed.

```python
from scipy.stats import mannwhitneyu

# List of numeric column names (excluding the timestamp column)
numeric_columns = [col for col in s32_df.columns if col != 'Timestamp']

# Convert the numeric columns to float data types
for column in numeric_columns:
    s32_df[column] = s32_df[column].astype(float)
    
for column in numeric_columns:
    s33_df[column] = s33_df[column].astype(float)

# Perform the Mann-Whitney U test for each numeric column
print("Mann-Whitney U test results:")
for column in numeric_columns:
    stat, p_value = mannwhitneyu(s32_df[column], s33_df[column], alternative='two-sided')
    result = {
        'Column': column,
        'Statistic': stat,
        'P-Value': p_value,
    }
    print(result)
```

 Null hypothesis states that there is no significant difference between the distributions of the two datasets (s32_df and s33_df)
 Alternate hypothesis states that there is a significant difference between the distributions of the two datasets (s32_df and s33_df)

For all columns, the p-values are smaller than 0.05, indicating strong evidence against the null hypothesis., which means that the p-values are highly statistically significant.
Based on these results, we would reject the null hypothesis for all columns. This means that there is a significant difference between the distributions of the two datasets (s32_df and s33_df) for all columns.

### Normality test between all dataframes

```python
import pandas as pd
from scipy import stats
import numpy as np

# Create a list of DataFrames
dataframes = [s32_df, S5_read_df, s18_video_df, s33_read_df, s33_video_df]

# Set the significance level (alpha)
alpha = 0.05

# Iterate through the DataFrames and perform the Shapiro-Wilk test for non-timestamp columns
for df in dataframes:
    # Extract the column names from the DataFrame
    column_names = df.columns
    
    # Exclude the timestamp column
    column_names = [col for col in column_names if col != 'timestamp']  # Replace 'timestamp' with your actual column name
    
    # Perform the Shapiro-Wilk test for each non-timestamp column
    for column in column_names:
        # Convert the column to a numeric data type (assuming it's not already)
        data = pd.to_numeric(df[column], errors='coerce')
        
        # Remove NaN values from the data
        data = data.dropna()
        
        # Check if the data has at least 3 data points
        if len(data) < 3:
            print(f"Skipping Shapiro-Wilk Test for '{column}' in DataFrame: Insufficient data (less than 3 data points)")
            continue
        
        statistic, p_value = stats.shapiro(data)
        
        # Print the results for each column
        print(f"Shapiro-Wilk Test for '{column}' in DataFrame:")
        print("Statistic:", statistic)
        print("P-value:", p_value)
        
        # Check if the p-value is less than alpha to determine normality
        if p_value < alpha:
            print("The data does not follow a normal distribution")
        else:
            print("The data follows a normal distribution")
        print("\n")
```

```python
from itertools import combinations
from scipy.stats import kruskal

# Create a dictionary to map dataframes to their names
dataframes = {
    's32_df': s32_df,
    'S5_read_df': S5_read_df,
    's18_video_df': s18_video_df,
    's33_read_df': s33_read_df,
    's33_video_df': s33_video_df
}

# Set the significance level (alpha)
alpha = 0.05

# Iterate through all combinations of dataframe names
for df1_name, df2_name in combinations(dataframes.keys(), 2):
    df1 = dataframes[df1_name]
    df2 = dataframes[df2_name]
    
    # Iterate through columns (excluding the "Timestamp" column)
    for column in df1.columns:
        if column != "Timestamp" and column in df2.columns:
            values1 = df1[column]
            values2 = df2[column]
            
            # Check if there is variance in the columns
            if values1.var() > 0 and values2.var() > 0:
                statistic, p_value = kruskal(values1, values2)
                
                # Print the dataframe names and column name
                print(f"Kruskal-Wallis Test between {df1_name} and {df2_name} for '{column}':")
                print("Statistic:", statistic)
                print("P-value:", p_value)
                
                # Check if the p-value is less than alpha to determine significance
                if p_value < alpha:
                    print("There are significant differences between the two datasets.")
                else:
                    print("There are no significant differences between the datasets.")
                print("\n")
            else:
                print(f"No variance in '{column}' between {df1_name} and {df2_name}. Skipping test.")

```

```python
from itertools import combinations
from scipy.stats import mannwhitneyu

# Create a dictionary to map dataframes to their names
dataframes = {
    's32_df': s32_df,
    'S5_read_df': S5_read_df,
    's18_video_df': s18_video_df,
    's33_read_df': s33_read_df,
    's33_video_df': s33_video_df
}

alpha = 0.05
significant_found = False

for df1_name, df2_name in combinations(dataframes.keys(), 2):
    df1 = dataframes[df1_name]
    df2 = dataframes[df2_name]
    
    combination_significant_found = False
    
    print(f"Comparison between {df1_name} and {df2_name}:")
    
    for column1 in df1.columns:
        if column1 != "Timestamp":
            for column2 in df2.columns:
                if column1 != "Timestamp" and column2 != "Timestamp":
                    values1 = df1[column1]
                    values2 = df2[column2]
                    
                    statistic, p_value = mannwhitneyu(values1, values2)
                    
                    if p_value < alpha:
                        print(f"Mann-Whitney U Test for '{column1}' in {df1_name} and '{column2}' in {df2_name}:")
                        print("Statistic:", statistic)
                        print("P-value:", p_value)
                        print("There is a significant difference between the two datasets.")
                        print("\n")
                        combination_significant_found = True
    
    if combination_significant_found:
        significant_found = True

if not significant_found:
    print("No statistically significant differences found between any combination of DataFrames and columns.")
```

```python
from itertools import combinations
from scipy.stats import mannwhitneyu

# Create a dictionary to map dataframes to their names
dataframes = {
    's32_df': s32_df,
    'S5_read_df': S5_read_df,
    's18_video_df': s18_video_df,
    's33_read_df': s33_read_df,
    's33_video_df': s33_video_df
}

alpha = 0.05
significant_found = False

for df1_name, df2_name in combinations(dataframes.keys(), 2):
    df1 = dataframes[df1_name]
    df2 = dataframes[df2_name]
    
    combination_significant_found = False
    
    print(f"Comparison between {df1_name} and {df2_name}:")
    
    # Get the common columns between df1 and df2
    common_columns = set(df1.columns).intersection(df2.columns)
    
    for column in common_columns:
        if column != "Timestamp":
            values1 = df1[column]
            values2 = df2[column]
            
            statistic, p_value = mannwhitneyu(values1, values2, alternative='greater')
            
            if p_value < alpha:
                print(f"Mann-Whitney U Test for '{column}' in {df1_name} and {df2_name}:")
                print("Statistic:", statistic)
                print("P-value:", p_value)
                print(f"{column} tends to be greater in {df1_name} than in {df2_name} with statistical significance.")
                print("\n")
                combination_significant_found = True
    
    if combination_significant_found:
        significant_found = True

if not significant_found:
    print("No statistically significant differences found between any combination of DataFrames and columns.")
```
